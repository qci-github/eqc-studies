This notebook includes the code to actually create the final plots in the paper (not including 3 dimensional visualisations), in cases where the numerical evaluation takes a considerable amount of time, data are loaded from a .npy file

```python
import numpy as np
import scipy.linalg as linalg
import scipy.sparse as sparse
import zeno_constraint as zc
import importlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
importlib.reload(zc)
H_proj_s_oh5=lambda s: zc.project_forbidden_weights(5,lambda x:x==1).todense()+1j*zc.multi_qubit_rotate_disp_ham(s*np.pi/2,5).todense()
project_precompute_oh5=zc.precompute_projections(H_proj_s_oh5)
```

```python
importlib.reload(zc)
Z=np.zeros([3,3])
Z[1,1]=1
Z[2,2]=-1
Z0=zc.kron_ith_position(Z,0,5).todense()
H_static=-2*zc.multi_var_u_offset_ham(5).todense()+Z0
final_gs=np.zeros(3**5)
final_gs[202]=1
start_state=np.zeros(3**5)
start_state[0]=1
t_tot_array=np.linspace(1,12,10)
subspace_eignvals=lambda H_sub: linalg.eigh(H_sub,eigvals_only=True)
z0_expect=lambda state_vec,project_vecs: np.dot(np.dot(np.dot(np.dot(state_vec.conj(),project_vecs.conj().T),Z0),project_vecs),state_vec)
final_subspace_prob=lambda state_vec,project_vecs: sum(abs(np.dot(np.dot(state_vec.conj(),project_vecs.conj().T),project_precompute_oh5[-1]))**2)
ground_state_prob=lambda state_vec,project_vecs: abs(np.dot(np.dot(state_vec.conj(),project_vecs.conj().T),final_gs))**2
measure_func=lambda state_vec,project_vecs,H_sub: [subspace_eignvals(H_sub),z0_expect(state_vec,project_vecs),final_subspace_prob(state_vec,project_vecs),ground_state_prob(state_vec,project_vecs)]
data_list_list_oh5=[None]*len(t_tot_array)
for i_t_tot in range(len(t_tot_array)):
    t_tot=t_tot_array[i_t_tot]
    data_list_list_oh5[i_t_tot]=zc.proj_space_sweep(H_proj_s_oh5,H_static,t_tot,start_state,measure_func,project_precompute=project_precompute_oh5)
```

```python
t_tot_array_many_oh=np.linspace(0,15,100)
final_probs_many_oh=np.zeros(len(t_tot_array_many_oh))
for i_t_tot in range(len(t_tot_array_many_oh)):
    #print(i_t_tot)
    t_tot=t_tot_array_many_oh[i_t_tot]
    data_many=zc.proj_space_sweep(H_proj_s_oh5,H_static,t_tot,start_state,measure_func,project_precompute=project_precompute_oh5)
    final_probs_many_oh[i_t_tot]=data_many[-1][3]
```

```python
fig=plt.figure()
ax0=plt.gca()
fig.add_axes([0.25,0.45,0.3,0.3])
ax1=plt.gca()
for i_t_tot in range(len(t_tot_array)):
    data_list=data_list_list_oh5[i_t_tot]
    final_state_prob_list=[]
    final_subspace_prob_list=[]
    for i_data in range(len(data_list)):
        final_state_prob_list=final_state_prob_list+[data_list[i_data][3]]
        final_subspace_prob_list=final_subspace_prob_list+[data_list[i_data][2]]
    line_color=[i_t_tot/len(t_tot_array),0,i_t_tot/len(t_tot_array)]
    ax0.plot(np.linspace(0,1/2,len(final_state_prob_list)),final_state_prob_list,color=line_color)
    ax0.plot(1/2,final_state_prob_list[-1],'o',color=line_color)
    ax1.plot(t_tot_array[i_t_tot],final_state_prob_list[-1],'o',color=line_color)
plt.sca(ax1)
plt.tick_params(axis='both', which='major', labelsize=12)
ax1.plot(t_tot_array_many_oh,final_probs_many_oh)
plt.ylim([0,1])
plt.xlim([0,15])
plt.xlabel('total runtime',fontsize=15)
plt.ylabel('final success',fontsize=15)
plt.sca(ax0)
ax0.plot(np.linspace(0,1/2,len(final_subspace_prob_list)),final_subspace_prob_list,'--k',linewidth=2)
plt.fill_between(np.linspace(0,1/2,len(final_subspace_prob_list)),final_subspace_prob_list,color='blue',alpha=0.2)
plt.xlim([0,1/2])
plt.ylim([0,1])
plt.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel(r'$\theta/\pi$',fontsize=15)
plt.ylabel('success probability',fontsize=15)
#plt.savefig('figures/oh5_success.pdf',bbox_inches='tight')
plt.show()
```
