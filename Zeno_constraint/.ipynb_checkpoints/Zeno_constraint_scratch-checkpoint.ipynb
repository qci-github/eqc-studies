```python
def double_arrow(head_1_coords,head_2_coords,ax_add,**kwargs):
    mp_coords=((head_1_coords[0]+head_2_coords[0])/2,(head_1_coords[1]+head_2_coords[1])/2)
    arrow1 = mpatches.FancyArrowPatch((mp_coords[0],mp_coords[1]),(head_1_coords[0],head_1_coords[1]),**kwargs)
    arrow2 = mpatches.FancyArrowPatch((mp_coords[0],mp_coords[1]),(head_2_coords[0],head_2_coords[1]),**kwargs)
    ax_add.add_patch(arrow1)
    ax_add.add_patch(arrow2)
```
